```python
data['model_new']=data['model_name'].str.extract(r'(\S{3,}|\S{1,2,}\s+\S+)')
#str-allows you to access the values of series as a string
#extract-used to extract in the regular expression as a string
#\s+ matches one or more white spaces
#\S{3,} matches three or more consecutive non-white space characters
#\S-matches one or more consecutive non white space characters
```

```python
data['model_new'].value_counts()
```

```python
data1=data.loc[(data['model_new']=='Swift')]
data2=data.loc[(data['model_new']=='Alto')]
data3=data.loc[(data['model_new']=='Wagon')]
```

```python
df=pd.concat([data1,data2,data3])
```

```python
df.head(6)
```

```python
df=df.drop(['model_name','city'],axis=1)
```

```python
df.head(4)
```

```python
df=pd.get_dummies(df)
```

```python
df.head(4)
```

```python
x=df.drop(['price (₹)'],axis=1)
```

```python
x.head(4)
```

```python
y=df['price (₹)']
```

```python
y.head(4)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
r=LinearRegression()
```

```python
reg=r.fit(x_train,y_train)
```

```python
y_pred=r.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred,y_test)
```

```python
Results=pd.DataFrame(columns=['price','predicted'])
Results['price']=y_test
Results['predicted']=y_pred
Results=Results.reset_index()
Results['ID']=Results.index
Results.head(15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='ID',y='price',data=Results.tail(50))
sns.lineplot(x='ID',y='predicted',data=Results.tail(50))
plt.plot()
```
